```python
import pandas as pd
import yaml
```

# In distribution

```python
all_mean_metrics = []
all_single_scores = []
for subset in ["CD4", "CD8"]:
    with open(f"/Users/alicedriessen/Box/LegacyFromOldColleagues/Alice/CAR_Tcells/Model/chemCPA/{subset}_ID_dummy_donor/cmonge_eval_test.yml") as stream:
        logs = yaml.safe_load(stream)
    for car, stats in logs.items():
        res = pd.Series(stats["mean_statistics"])
        res = res.reset_index(drop=False).rename(columns={"index": "metric", 0: "score"})
        res["embedding"] = "ESM XL"
        res["eval_car"] = car
        res["eval_setting"] = "chemCPA"
        res["subset"] = subset
        all_mean_metrics.append(res)

        single_scores = pd.DataFrame({"r2": stats["r2"], "mmd": stats["mmd"]})
        single_scores["embedding"] = "ESM XL"
        single_scores["eval_car"] = car
        single_scores["eval_setting"] = "chemCPA"
        single_scores["subset"] = subset
        all_single_scores.append(single_scores)
```

```python
results = pd.concat(all_mean_metrics)
single_scores = pd.concat(all_single_scores)
results
```

```python
single_scores
```

```python
results.to_csv("/Users/alicedriessen/Box/LegacyFromOldColleagues/Alice/CAR_Tcells/Model/chemCPA/ID_CD4_CD8_carot_results.csv")
single_scores.to_csv("/Users/alicedriessen/Box/LegacyFromOldColleagues/Alice/CAR_Tcells/Model/chemCPA/ID_CD4_CD8_carot_single_scores.csv")
```

# Model per car

```python
with open("/Users/alicedriessen/Box/LegacyFromOldColleagues/Alice/CAR_Tcells/Model/OT/CAR_variants.txt", "r") as f:
    cars = f.read().splitlines()
cars
```

```python
all_mean_metrics = []
for subset in ["CD4", "CD8"]:
    for car in cars:
        with open(f"/Users/alicedriessen/Box/LegacyFromOldColleagues/Alice/CAR_Tcells/Model/chemCPA/model_per_car/{subset}/{car}/cmonge_eval_test.yml") as stream:
            logs = yaml.safe_load(stream)
        for car_log, stats in logs.items():
            if car_log == "NA-NA-NA":
                continue
            res = pd.Series(stats["mean_statistics"])
            res = res.reset_index(drop=False).rename(columns={"index": "metric", 0: "score"})
            res["CAR"] = car_log
            res["setting"] = "chemCPA"
            res["subset"] = subset
            all_mean_metrics.append(res)
```

```python
results = pd.concat(all_mean_metrics)
results
```

```python
results.to_csv("/Users/alicedriessen/Box/CAR_Tcells/Model/chemCPA/model_per_car/ID_CD4_CD8_carot_results.csv")
```
